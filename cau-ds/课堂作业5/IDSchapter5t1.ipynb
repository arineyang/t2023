```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

filename="D:\mine_coding\CAU_IDS\data5\wine.csv"
data=pd.read_csv(filename,index_col='idx') #data #index_col=0
data[:5]
```

```python
# 1.使用scikit-learn建立SVM模型为葡萄酒数据集构造分类器（分类结果为’good’或‘not’ ） (40分)[“不可使用quantity”列]
# 2.评估分类器在此数据集上的分类性能(20分)
# 	使用SVM多分类的时候需要用到Kappa系数
# 	* 需要划分训练集和测试集

#替换good not为10
data.loc[data['good_or_not']=='good','good_or_not']=1
data.loc[data['good_or_not']=='not','good_or_not']=0
data[:5]
```

```python
#特征和标签
x = data.drop(['good_or_not','quality'], axis=1).values.astype(float)
y = data['good_or_not'].values.astype(float)
print(type(x),type(y))
```

```python
from sklearn import preprocessing
from sklearn import model_selection
from sklearn import metrics
from sklearn import svm

# 对不同方差的数据标准化
x_scale = preprocessing.scale(x)

# 将标准化后的数据集拆分为训练集和测试集，在测试集上查看分类效果
x_train, x_test,  y_train, y_test = model_selection.train_test_split(x_scale, y, test_size=0.2,random_state=1)

# 使用scikit-learn建立SVM模型为葡萄酒数据集构造分类器
clf = svm.SVC(kernel='rbf',gamma=0.4, C=1)
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
# 查看在测试集上混淆矩阵，Precision、Recall和F1
a = clf.score(x_test, y_test)
print('Accuracy: ', a)
print(metrics.classification_report(y_test, y_pred))
print('Confusion Matrix:')
print(metrics.confusion_matrix(y_test, y_pred))
```
