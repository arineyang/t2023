```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
from pandas import Series

fig=plt.figure(figsize=(8,6))
pcdi = [1.47,1.62,1.78,1.94,2.38,2.60,2.82,3.07,3.21]
data1 = DataFrame({'PCDI: wanyuan':pcdi},index=['2012','2013','2014','2015','2016','2017','2018','2019','2020'])
#data1.plot()
#plt.show()
data1.plot(title='Income',linewidth=2, marker='s', linestyle='dashed',color='k', grid=True,alpha=0.9,use_index=True) # linestyle='dashed'线型虚线 ylim=(35,75)
plt.annotate('peak',xy=(8,3.21),xytext=(8.5,3.25), arrowprops=dict(arrowstyle='->'))
plt.xlabel('YEAR')
plt.ylabel('PCDI')
#plt.show()

ax1=fig.add_subplot(2,2,1) #子图1折线图
ax1.plot(data1)
ax2=fig.add_subplot(2,1,2) #子图2箱形图
data1.plot(kind='box',fontsize='small',xticks=[],ax=ax2)
ax3=fig.add_subplot(2,2,2) #子图3垂直柱状图
data1.plot(kind='bar',use_index=True,fontsize='small',ax=ax3)

plt.subplots_adjust(wspace = 0.3, hspace = 0.3) #left=0.2, bottom=0.1, right=0.8, top=0.9
#plt.show()

#figure.savefig(filename,dpi,bbox_inches) 
#plt.savefig(filename,dpi,bbox_inches)
plt.savefig('2012-2020PCDI.jpg',dpi=400,bbox_inches='tight')
```
